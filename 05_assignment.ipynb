```python
import pandas as pd

path = 'assignment_data/test_1.csv' # do check file path before running the cell
data = pd.read_csv(path, engine='python', on_bad_lines='skip') # on_bad_lines is used because read_csv was giving error
data.head()
```

```python
data.columns
```

Extracting date and time from timestamp column. Timestamp column also contain timezone, so below we will separate timezone from datetime.

```python
data[['datetime', 'timezone']] = data['timestamp'].str.rsplit('.', n=1, expand=True)
data[['datetime', 'timezone']].head()
```

```python
data['date'] = pd.to_datetime(data['datetime']).dt.date
data['date'][:5]
```

```python
data['time'] = pd.to_datetime(data['datetime']).dt.time
data['time'][:5]
```

Now, since we have separate date and time column, we don't need timezone and datetime column. So we will delete them from dataframe.

```python
data.drop(['datetime', 'timezone'], axis=1, inplace=True)
```
